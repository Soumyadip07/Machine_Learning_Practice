```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=10)
```

```python
X_train
```

```python
y_train
```

```python
#Standarization or feature scaling of the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression
```

```python
regression.fit(X_train,y_train)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

### Assumptions in linear regresion

```python
plt.scatter(y_test,reg_pred)
```

```python
## residual
residuals=y_test-reg_pred
```

```python
residuals
```

```python
sns.displot(residuals,kind='kde')
```

```python
## scatter plot between predictions and residual
```

```python
plt.scatter(reg_pred,residuals)
```

```python
#performance metrics
from sklearn.metrics import mean_squared_error,mean_absolute_error
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
## R square and adjusted R square
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

```python
## adjusted r square
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
ridge=Ridge()
```

```python
ridge.fit(X_train,y_train)
```

```python
print(ridge.coef_)
```

```python
print(ridge.intercept_)
```

```python
ridge_pred=ridge.predict(X_test)
```

```python
ridge_pred
```

### Assumption

```python
plt.scatter(y_test,ridge_pred)
```

```python
residuals=y_test-ridge_pred
sns.displot(residuals,kind='kde')
```

```python
plt.scatter(ridge_pred,residuals)
```

```python
print('MSE:- ',mean_squared_error(y_test,ridge_pred))
print("MAE:- ",mean_absolute_error(y_test,ridge_pred))
print("RMSE:- ",np.sqrt(mean_squared_error(y_test,ridge_pred)))
```

```python
score=r2_score(y_test,ridge_pred)
print("R2 Score:- ",score)
```

```python
adj_r2=1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
print("Adjusted R2:- ",adj_r2)
```
